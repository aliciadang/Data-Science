#  Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
result = pd.read_sql("SELECT M.league_id, L.name as league_name, COUNT(M.id) AS number_of_matches FROM Match M, League as L WHERE M.league_id = L.id GROUP BY league_id ORDER BY COUNT(M.id) DESC", conn)
result
```

#  Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu

```python
result = pd.read_sql("SELECT M.league_id, L.name as league_name, COUNT(M.id) AS number_of_matches, SUM(M.home_team_goal) as sum_home_team_goal, sum(M.away_team_goal) as sum_away_team_goal FROM Match M, League as L WHERE M.league_id = L.id GROUP BY league_id ORDER BY COUNT(M.id) DESC", conn)
result
```

#  Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất

```python
result = pd.read_sql("""SELECT T.team_api_id, T.team_long_name, SUM(M.home_team_goal) as sum_home_team_goal
                        FROM Team as T, Match as M
                        WHERE T.team_api_id = M.home_team_api_id 
                        GROUP BY T.team_api_id 
                        ORDER BY sum_home_team_goal DESC LIMIT 20""", conn)
result
```

#  Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất

#  Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp theo số lượng giảm dần

```python
result = pd.read_sql("""SELECT T.team_long_name, SUM(goals) AS total_goals
                        FROM
                            (SELECT home_team_api_id AS team_api_id, SUM(home_team_goal) goals
                            FROM match
                            GROUP BY home_team_api_id
                            UNION
                            SELECT away_team_api_id AS team_api_id, SUM(away_team_goal) goals
                            FROM match
                            GROUP BY away_team_api_id)
                        AS TG
                        INNER JOIN TEAM T 
                        ON T.team_api_id = TG.team_api_id
                        GROUP BY T.team_api_id
                        ORDER BY total_goals DESC""", conn)
result
```

#  Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
result = pd.read_sql("""SELECT TM.team_api_id, T.team_long_name, SUM(TM.matches) AS total_matches
                        FROM
                           (SELECT home_team_api_id AS team_api_id, COUNT(*) AS matches
                            FROM Match
                            GROUP BY team_api_id
                            UNION 
                            SELECT away_team_api_id AS team_api_id, COUNT(*) AS matches
                            FROM Match
                            GROUP BY team_api_id)
                        AS TM, Team AS T
                        WHERE T.team_api_id = TM.team_api_id
                        GROUP BY TM.team_api_id
                        ORDER BY total_matches DESC""", conn)
result
```

#  Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội

```python
result = pd.read_sql("""SELECT T.team_api_id, T.team_long_name,                    
                            SUM(CASE WHEN home_team_goal > away_team_goal THEN 1
                                 ELSE 0 END) AS home_wins,
                            SUM(CASE WHEN home_team_goal = away_team_goal THEN 1
                                 ELSE 0 END) AS home_draws,
                            SUM(CASE WHEN home_team_goal < away_team_goal THEN 1
                                 ELSE 0 END) AS home_losses
                        FROM Match As M     
                        INNER JOIN Team AS T
                        ON M.home_team_api_id = T.team_api_id
                        GROUP BY T.team_long_name""", conn)
result
```

#  Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
result = pd.read_sql("""SELECT T.team_api_id, T.team_long_name,                    
                            SUM(CASE WHEN home_team_goal > away_team_goal THEN 1
                                 ELSE 0 END) AS home_wins
                        FROM Match As M     
                        INNER JOIN Team AS T
                        ON M.home_team_api_id = T.team_api_id
                        GROUP BY T.team_long_name
                        ORDER BY home_wins DESC
                        LIMIT 10""", conn)
result
```

# Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
result = pd.read_sql("""SELECT TM1.league, TM1.number_of_matches, ROUND(TM1.number_of_matches*100.0/TM2.total_matches,2) AS percentage
                        FROM
                            (SELECT M.league_id, L.name AS league, COUNT(*) AS number_of_matches 
                             FROM Match M, League AS L 
                             WHERE M.league_id = L.id 
                             GROUP BY league_id) AS TM1, 
                            (SELECT COUNT(*) AS total_matches FROM Match) AS TM2
                        GROUP BY TM1.league
                        ORDER BY TM1.number_of_matches DESC""", conn)
result
```

#  Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận

```python
result = pd.read_sql("""SELECT TM3.league_id, TM3.league, TM3.goal, ROUND((TM3.goal*100.0/TM4.total_goal),2) AS percentage
                        FROM
                        (SELECT M.league_id, L.name as league, (SUM(M.home_team_goal + M.away_team_goal)) AS goal
                        FROM Match M, League AS L 
                        WHERE M.league_id = L.id 
                        GROUP BY league_id) AS TM3,
                        (SELECT (SUM(home_team_goal + away_team_goal)) AS total_goal FROM Match) AS TM4
                        GROUP BY league_id
                        ORDER BY TM3.goal DESC""", conn)
result
```

#  Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải

```python
result = pd.read_sql("""SELECT M.league_id, L.name as league, M.season, (SUM(M.home_team_goal + M.away_team_goal)) AS goal
                        FROM Match M, League AS L 
                        WHERE M.league_id = L.id
                        GROUP BY league_id, season""", conn)
result
```

# Câu hỏi 19: Liệt kê thông tin cầu thủ:
* Quy đổi cân nặng sang kilogram
* Quy đổi chiều cao sang mét
* Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)
* Tính tuổi của cầu thủ

```python
result = pd.read_sql("""SELECT id, player_api_id, player_name, player_fifa_api_id, birthday, 
                            ROUND(height/100,2) AS height, 
                            ROUND(weight*0.453592) AS weight,
                            ROUND(((weight* 0.453592) / (height/100 * height/100)),1) AS bmi,
                            (current_date - birthday) AS age
                        FROM Player""", conn)
result
```

# Câu hỏi 20: Liệt kê cầu thủ lớn tuổi nhất

```python
result = pd.read_sql("""SELECT id, player_api_id, player_name, player_fifa_api_id, birthday, 
                            ROUND(height/100,2) AS height, 
                            ROUND(weight*0.453592) AS weight,
                            ROUND(((weight* 0.453592) / (height/100 * height/100)),1) AS bmi,
                            MAX(current_date - birthday) AS age
                        FROM Player""", conn)
result
```

# Câu hỏi 21: Liệt kê cầu thủ có số trận đấu nhiều nhất

```python
result = pd.read_sql("""SELECT player_api_id, player_fifa_api_id, player_name, num_of_matches
                        FROM
                            (SELECT PA.player_api_id, PA.player_fifa_api_id, P.player_name, COUNT(PA.id) AS num_of_matches
                            FROM Player_Attributes AS PA, Player AS P
                            WHERE PA.player_api_id = P.player_api_id 
                            GROUP BY PA.player_api_id)
                        GROUP BY player_api_id
                        ORDER BY num_of_matches DESC
                        LIMIT 2
                     """, conn)
result
```

# Câu hỏi 22: Liệt kê các cầu thủ có tổng số điểm đánh giá lớn hơn 80

```python
result = pd.read_sql("""SELECT PA.player_api_id, PA.player_fifa_api_id, P.player_name, PA.overall_rating
                        FROM Player_Attributes AS PA, Player AS P
                        WHERE PA.player_api_id = P.player_api_id 
                        GROUP BY PA.player_api_id
                        HAVING overall_rating > 80
                        ORDER BY overall_rating DESC
                     """, conn)
result
```
