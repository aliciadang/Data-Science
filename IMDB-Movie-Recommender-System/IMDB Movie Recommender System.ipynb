## Task 1: Đọc dữ liệu creadits và movies và kết hợp 2 dữ liệu thành một dataframe

```python
df1 = pd.read_csv('credits.csv')
df2 = pd.read_csv('movies.csv')
```

# Demographic Filtering

## Task 2: Xây dựng hàm tính weighted_rating

```python
# Tính giá trị C
C = df['vote_average'].mean()
C
```

```python
# Tính giá trị m
m = df['vote_count'].quantile(0.9)
m
```

```python
# Xây dựng hàm tính weighted_rating
def weighted_rating(df, m=m, C=C):
    v = df['vote_count']
    R = df['vote_average']
    return (v/(v+m)*R + m/(v+m)*C)
```

## Task 3: Tính điểm cho từng bộ phim từ hàm weighted_ratings và đề xuất phim

```python
# Lọc những bộ phim có số lượng đánh giá ít nhất bằng m
q_movies = df.loc[df['vote_count'] >= m]
q_movies.shape
```

```python
q_movies.head()
```

```python
# Tính điểm cho từng bộ phim bằng hàm weighted_rating
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head()
```

```python
# Đề xuất 10 bộ phim dựa vào phương pháp đề xuất Demographic Filtering
q_movies.sort_values(by='score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

## Task 4: Trực quan biểu đồ trực quan 10 bộ phim phổ biến nhất hiện nay bằng chỉ số popularity.

```python
top_popularity = q_movies.sort_values(by='popularity', ascending=False)
top_popularity = top_popularity[['title', 'popularity']].head(10)
top_popularity
top_popularity.plot(kind='barh', figsize=(10,4), x='title', y='popularity', title='Popular Movies', legend=None)
plt.xlabel('Popularity')
plt.ylabel('')
plt.show()
```

## Task 5: Tính IF-IDF cho toàn bộ overview

```python
# Kiểm tra các dữ liệu NA
df['overview'].isna().sum()
```

```python
# Xử lý dữ liệu NA
df['overview'] = df['overview'].fillna("")
```

```python
# Tính IF-IDF cho toàn bộ overview
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['overview'])
tfidf_matrix.shape
```

## Task 6: Tính ma trận tương tự cho overview từ TF-IDF

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

## Task 7: Xây dựng hàm để đề xuất phim dựa vào tổng quan của phim

```python
# Tạo Series với value là index, index là title của bộ phim
overview_series = pd.Series(data=df.index, index=df['title'])
overview_series
```

```python
# Xây dựng hàm đề xuất phim
def get_recommendations(df, title, cosine_sim=cosine_sim):
    # Lấy index của bộ phim thông qua title
    get_index = overview_series[title]
    
    # Lấy danh sách giá trị tương đồng của bộ phim cụ thể
    similarity_list = list(enumerate(cosine_sim[get_index]))
    
    # Sắp xếp danh sách trên theo giá trị tương tự giảm dần
    similarity_list = sorted(similarity_list, key=lambda x: x[1], reverse=True)
    
    # Lấy ra 10 phần tử đầu từ danh sách trên
    similarity_list = similarity_list[1:11]
    top10_index = (i[0] for i in similarity_list)
    
    # Trả về tiêu đề theo index có trong danh sách 10 bộ phim mà ta muốn khuyến nghị
    return df['title'].iloc[top10_index]
```

```python
get_recommendations(df, 'The Dark Knight Rises')
```

```python
get_recommendations(df, 'The Avengers')
```

## Task 8: Lấy ra tên tác giả từ cột crew và lưu vào cột director

```python
# Xây dựng hàm lấy tên tác giả từ cột crew
def director_name(a):
    for i in a:
        if i['job'] == 'Director':
            return i['name']
    return np.nan        
```

```python
# Lưu các kết quả vào cột mới 'director'
df['director'] = df['crew'].apply(director_name)
df['director']
```

## Task 9: Lấy ra 3 diễn viên đầu tiên từ cột cast và lưu vào chính cột cast. Làm điều tương tự với keywords và genres

```python
# Xây dựng hàm lấy ra danh sách top 3 thỏa yêu cầu
def top3_list(a):
    if isinstance(a, list):
        list1 = [i['name'] for i in a]
        if len(list1) >= 3:
            list1 = list1[:3]
        return list1
    return []
```

```python
# Áp dụng hàm đã xây dựng cho 3 cột sau, lưu kết quả vào chính cột đó
features2 = ['cast', 'keywords', 'genres']
for i in features2:
    df[i] = df[i].apply(top3_list)
```

```python
df[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

## Task 10: Điều chỉnh dữ liệu phù hợp với đầu vào của thuật toán

```python
# Xây dựng hàm điều chỉnh dữ liệu phù hợp với đầu vào của thuật toán
def fixed_data(a):
    if isinstance(a, list):
        return [str.lower(i.replace(' ', '')) for i in a]
    else:
        if isinstance(a, str):
            return str.lower(a.replace(' ', ''))
        else:
            return ''
```

```python
# Áp dụng hàm đã xây dựng cho các cột sau
features3 = ['keywords', 'cast', 'director', 'genres']
for i in features3:
    df[i] = df[i].apply(fixed_data)
```

```python
df[features3].head()
```

## Task 11: Kết hợp dữ liệu của từng giá trị trong keywords, cast, director, genres thành một chuỗi và lưu vào cột movie_info

```python
# Xây dựng hàm kết hợp dữ liệu thành 1 chuỗi
def movie_info(a):
    return ' '.join(a['keywords']) + ' ' + ' '.join(a['cast']) + ' ' + a['director'] + ' ' + ' '.join(a['genres'])

# Áp dụng kết quả cho các cột trên và lưu vào cột mới
df['movie_info'] = df.apply(movie_info, axis=1)
```

```python
df['movie_info'].head()
```

## Task 12: Tính CountVectorizer cho từng movie_info

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['movie_info'])
```

## Task 13: Tính ma trận tương tự cho movie_info từ CountVectorizer

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Tạo Series với value là index, index là title của bộ phim
movieinfo_series = pd.Series(data=df.index, index=df['title'])

# Xây dựng hàm đề xuất phim mới
def get_recommendations(title, cosine_sim=cosine_sim2):
    # Lấy index của bộ phim thông qua title
    get_index = movieinfo_series[title]
    
    # Lấy danh sách giá trị tương đồng của bộ phim cụ thể
    similarity_list = list(enumerate(cosine_sim2[get_index]))
    
    # Sắp xếp danh sách trên theo giá trị tương tự giảm dần
    similarity_list = sorted(similarity_list, key=lambda x: x[1], reverse=True)
    
    # Lấy ra 10 phần tử đầu từ danh sách trên
    similarity_list = similarity_list[1:11]
    top10_index = (i[0] for i in similarity_list)
    
    # Trả về tiêu đề theo index có trong danh sách 10 bộ phim mà ta muốn khuyến nghị
    return df['title'].iloc[top10_index]
```

```python
get_recommendations('The Dark Knight Rises', cosine_sim2)
```

```python
get_recommendations('The Godfather', cosine_sim2)
```
